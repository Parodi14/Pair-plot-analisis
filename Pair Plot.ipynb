# Pair Plot

## Es otro gráfico compuesto.

Es otro gráfico compuesto. Es un gráfico de "todos contra todos". Compara al mismo tiempo todas las variables numéricas de mi dataset a través de diversos Scatter plots arreglados en un grid (N x N). Siendo N el número de variables en mi dataset. En cada fila del grid, se compara en diversos scatter plot, una variable numérica VERSUS todas las demás. La diagonal del grid va a tener un histograma o un KDE de esa variable. (Ya que no tendría utilidad hacer un scatter plot donde ambos ejes correspondan a la misma variable)

```python
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('pinguinos.csv')
df = df.dropna()
df 
```

```python
sns.pairplot(
    data=df,
    #hue= 'species',
    #kind = 'hist'
)
```

```python
sns.pairplot(
    data=df,
    hue= 'island'

)
```
